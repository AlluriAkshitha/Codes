```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace the path with the actual file path)
df = pd.read_csv('Day_11_banking_data.csv')

# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
transaction_sum_per_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Bar plot
transaction_sum_per_account_type.plot(kind='bar', color='skyblue', title='Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
transaction_per_branch = df['Branch'].value_counts()

# Pie chart
transaction_per_branch.plot(kind='pie', autopct='%1.1f%%', title='Percentage of Transactions per Branch', figsize=(7, 7))
plt.ylabel('')  # Hide the y-label for better visualization
plt.show()
```
