```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Day_13_Pharma_data.csv')

# 1. Bar plot comparing the average Effectiveness for each drug across different regions
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product_Name', y='Effectiveness', hue='Region')
plt.title('Average Effectiveness for Each Drug Across Different Regions')
plt.xlabel('Product')
plt.ylabel('Average Effectiveness')
plt.show()

# 2. Violin plot to show the distribution of Effectiveness and Side_Effects for each product
# Ensure 'Side_Effects' has exactly two levels for split
df['Side_Effects'] = df['Side_Effects'].apply(lambda x: 'Positive' if x == 'None' else 'Negative')

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Product_Name', y='Effectiveness', hue='Side_Effects', split=True)
plt.title('Distribution of Effectiveness and Side Effects for Each Product')
plt.xlabel('Product')
plt.ylabel('Effectiveness')
plt.show()

# 3. Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']], hue='Side_Effects')
plt.suptitle('Pairplot: Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# 4. Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Trial_Period', y='Effectiveness')
plt.title('Effectiveness Comparison for Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# 5. Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Marketing_Spend', y='Effectiveness', scatter_kws={'s': 20, 'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Regression: Marketing Spend vs Drug Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
